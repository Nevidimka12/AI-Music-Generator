```python
import pandas as pd
df = pd.read_csv('TestLibrossa.csv', index_col = "Unnamed: 0")
tuples = [tuple(x) for x in df.values]
dicts = df.to_dict().values()
```

```python
import os
import numpy as np
np.set_printoptions(precision=4)
import catboost
from catboost import *
from catboost import datasets
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
df.shape
```

```python
X_train , X_validation ,y_train, y_validation = train_test_split(df.drop('target', axis = 1), df['target'], random_state=13, test_size=0.1)
```

```python
from catboost import CatBoostClassifier
params = {'iterations': 3000,
          'random_state': 48,
          'depth': 7,
          'learning_rate': 0.03,
          'task_type':"GPU",
          'save_snapshot':True,
          'snapshot_file':'shapshot.bkp', 
          'snapshot_interval':1
          }
model = CatBoostClassifier(**params)
model.fit(
    X_train, y_train,
    eval_set=(X_validation, y_validation),
    use_best_model=True
)
```

```python
model.save_model("Librosa_model2")
```
